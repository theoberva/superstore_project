```python
%matplotlib inline

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('superstore.csv',encoding='windows-1254')
df.sample(n=10)
```

```python
print(df.info())

## Convert date columns into datetime dtype

df['Order Date'] = pd.to_datetime(df['Order Date'],format="%m/%d/%Y")
df['Ship Date'] = pd.to_datetime(df['Ship Date'],format="%m/%d/%Y")

print(df.info())
```

```python
df.describe()
```

```python
#check for any missing values

df.isna().sum()
```

```python
# Check for patterns and relationships

sns.heatmap(df.corr(),annot=True)
plt.show() # strong relationships with profit/sales & discoutn/profit

```

```python
#see sales growth over the years

sales_year = df.groupby(df['Order Date'].dt.year)['Sales'].sum()
sales_year.plot.bar()
plt.show()
   
```

```python
# Look at the profits comapred to the sales

sales_mth_yr = df.groupby([df['Order Date'].dt.year,df['Order Date'].dt.month ])['Sales'].sum()
profit_mth_yr = df.groupby([df['Order Date'].dt.year,df['Order Date'].dt.month ])['Profit'].sum()

ax, fig = plt.subplots(figsize = (10,7))
sales_mth_yr.plot.bar()
profit_mth_yr.plot.bar(color='Orange')
ax.legend()
plt.show()
```

```python
# Compare sales and profit in diferent cities

top10_city_sales = df.groupby('City')['Sales'].sum().sort_values(ascending=False).head(n=10)
top10_city_profits = df.groupby('City')['Profit'].sum().sort_values(ascending=False).head(n=10) 

top10_city_sales.plot.bar()
top10_city_profits.plot.bar(color='orange')
plt.legend()
plt.show()
```

```python
# Compare sales and profit in diferent regions


region_sales = df.groupby('Region')['Sales'].sum().sort_values(ascending=False).head(n=10)
region_profits = df.groupby('Region')['Profit'].sum().sort_values(ascending=False).head(n=10) 

region_sales.plot.bar()
region_profits.plot.bar(color='orange')
plt.legend()
plt.show()
```

```python
# Top sales by product

top10_products = df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(n=10)

fig, ax = plt.subplots(figsize=(10,7))

ax.barh(top10_products.index,top10_products.values)
ax.invert_yaxis()
plt.show()
```

```python
# Top sale/profit by category

top_category_sales = df.groupby('Category')['Sales'].sum().sort_values(ascending=False)
top_category_profit = df.groupby('Category')['Profit'].sum().sort_values(ascending=False)

fig, ax = plt.subplots(figsize=(8,5))

ax.barh(top_category_sales.index, top_category_sales.values)
ax.barh(top_category_profit.index, top_category_profit.values)
ax.invert_yaxis()
plt.show()
```

```python
#top sub-category

top_subcategory = df.groupby('Sub-Category')['Sales'].sum().sort_values(ascending=False)
top_subcategory_profit = df.groupby('Sub-Category')['Profit'].sum().sort_values(ascending=False)

fig, (ax1,ax2) = plt.subplots(1,2, figsize=(15,8))

ax1.barh(top_subcategory.index, top_subcategory.values)
ax2.barh(top_subcategory_profit.index, top_subcategory_profit.values)
ax1.title.set_text('Sales')
ax2.title.set_text('Profits')
ax1.invert_yaxis()
ax2.invert_yaxis()
plt.show()
```

```python
# Top sales by segment

top_segment = df.groupby('Segment')['Sales'].sum().sort_values(ascending=False)
top_segment_profit = df.groupby('Segment')['Profit'].sum().sort_values(ascending=False)


fig, ax = plt.subplots(figsize=(8,5))

ax.barh(top_segment.index, top_segment.values)
ax.barh(top_segment_profit.index, top_segment_profit.values)
ax.invert_yaxis()
plt.show()
```

```python
#average sales per order

print('Avg. value per order: ', df.groupby('Order ID')['Sales'].mean().mean())
print('Avg. profit per order: ', df.groupby('Order ID')['Profit'].mean().mean())
print('Avg. profit % per order: ',round(((df.groupby('Order ID')['Profit'].mean().mean())
                                         / (df.groupby('Order ID')['Sales'].mean().mean())) * 100,2))

sns.kdeplot(data=df.groupby('Order ID').sum(), x='Sales') #the distribution of sales per order
plt.show()
```

```python
# average discount 

print('Avg. discount % per order: ', round(df.groupby('Order ID')['Discount'].mean().mean() *100,2))
```

```python
df.groupby(['Customer ID', 'Customer Name']).agg({'Order ID':'count','Sales':'sum', 'Quantity':'sum', 'Discount':'mean','Profit':'sum'}).sort_values(by='Sales',ascending = False).head(15)
#Top customers that spend the most overall, average discount given, how many orders theyve made.
## Our customer with the most sales has almost cost us $2000
```
